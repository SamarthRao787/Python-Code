```python
#Assigning loyalty flag
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['department_id','order_hour_of_day','product_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] < 10 , 'spender_type'] = 'Low Spender'
```

```python
#Creating a flag for frequent and non frequent users
```

```python
ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge['median_prior_order_time'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_order_time'] > 20, 'frequency_flag'] = 'non_frequent_customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_prior_order_time'] >10) | (ords_prods_merge['median_prior_order_time'] <=20), 'frequency_flag'] = 'regular_customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_order_time'] <= 10, 'frequency_flag'] = 'frequent_customer'
```
